```python
import pandas as pd
```

---

```python
# Store all flight dataset paths
files = [f'data/us_flights_2024_{i}.csv' for i in range(1, 13)]

# Reading and concat each DF
df_all = pd.concat((pd.read_csv(f) for f in files), ignore_index=True)

# Pick only SDF airport traffic
df_all_sdf = df_all[(df_all['Origin'] == 'SDF') | (df_all['Dest'] == 'SDF')].copy()
```

```python
# Save as Parquet to reduce file size
df_all.to_parquet('data/us_flights_2024_all.parquet')
df_all_sdf.to_parquet('data/us_flights_2024_all_sdf.parquet')
```

```python
df_all.to_csv('data/us_flights_2024_all.csv')
df_all_sdf.to_csv('data/us_flights_2024_all_sdf.csv')
```
